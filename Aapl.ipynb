```python
import pandas as pd
import numpy as np
import yfinance as yf
import seaborn as sns
```

```python
aapl = yf.download('AAPL',
                   start='2019-04-02',
                   end='2025-04-08',
                   progress=False)
```

```python
import matplotlib.pyplot as plt
aapl['Close'].plot(figsize=(8,5)).axhline(y=230, color='r' , linestyle='--', label='Reference line')
plt.legend() # display the legend
plt.show()
```

```python
aapl['close_simple_rtn'] = aapl['Close'].pct_change()

```

```python
aapl['close_simple_rtn'].plot(figsize=(8,5),color='g').axhline(y=0, color='r' , linestyle='--', label='Reference line')
plt.legend() # display the legend
plt.show()
```

```python
aapl['close_simple_rtn'].describe()
```

```python
aapl['close_simple_rtn'].corr(aapl['close_simple_rtn'].shift(12))
```

```python
# calculate skewness and kurtosis
skewness = aapl['close_simple_rtn'].skew()
kurtosis = aapl['close_simple_rtn'].kurtosis()

print("Skewness: ", skewness)
print("Kurtosis: ", kurtosis)
```
